### Line Plots

```python
import numpy as np
import matplotlib.pyplot as plt

x = np.arange(10)
print(x)

y1 = x**2
print(y1)

y2 = 2*x + 3
print(y2)

plt.plot(x,y1)
plt.show()

themes = plt.style.available
print(themes)

plt.style.use("seaborn")
plt.plot(x,y1,color='red',label="Apple",linestyle='dashed')
plt.plot(x,y2,color='green',label='Kiwi',marker='o')
plt.xlabel('Time')
plt.ylabel('Prices')
plt.title('Graph of prices of fruits with time')
plt.legend()
plt.show()
```

### Scatter Plots

```python
plt.scatter(x,y1,marker='*')
plt.scatter(x,y2,color='red')
plt.figure(figsize=(20,10))
plt.show()
```

### Bar Graphs

```python
plt.bar([0,1,2],[10,20,50])
plt.show()
```

```python
x_coordinates = np.array([0,1,2])*2
plt.bar(x_coordinates,[10,20,30],width=0.5,label='current year',tick_label=["Gold","Platinum","Silver"])
plt.bar(x_coordinates+0.5,[40,50,60],width=0.5,label='next year')
plt.title("Metal price comparison")
plt.xlabel("Metal")
plt.ylabel("Price")
plt.legend()
plt.ylim(0,100)
#plt.style.use("dark_background")
plt.show()
```

### Pie Charts

```python
subjects = ["Maths","Chem","Physics","English"]
weightage = [20,10,15,5]
plt.pie(weightage,labels=subjects)
plt.show()
```

```python
plt.pie(weightage,labels=subjects,explode=(0,0,0.1,0),shadow='true')
plt.style.use("seaborn")
plt.show()
```
